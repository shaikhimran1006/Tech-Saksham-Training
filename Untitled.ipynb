```python
df=sns.load_dataset('tips')
```

```python
df.head(10)
```

```python
plt.figure(figsize=(3,3))
sns.barplot(x='time',y='total_bill',data=df)
```

```python
plt.figure(figsize=(3,3))
sns.barplot(x='time',y='total_bill',hue='sex',data=df,palette='coolwarm')
```

```python
plt.Figure(figsize=(4,40))
sns.barplot(x="time",y="total_bill",data=df,hue="sex",palette="coolwarm",orient="h")
plt.show()
```

```python
#histogram
sns.distplot(df["total_bill"],kde=False,color="red")
plt.grid()
plt.show()
```

```python
#histogram
sns.distplot(df["total_bill"])
plt.grid()
plt.show()
```

```python
#scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df,hue="sex",style="sex",markers=["s","D"])
plt.show()
```

```python
#scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df,hue="sex",style="sex",markers=["s","D"],palette="dark")
plt.show()
```

```python
#boxplot
sns.boxplot(x="time",y="tip",data=df,color="red")
plt.show()
```

```python
#violinplot
sns.violinplot(x="time",y="tip",data=df,hue="sex",inner="quarter",split=True)
plt.show()
```

```python
plt.figure(figsize=(2,2))
a=np.array([[1,20,30],[10,20,3],[1,2,3]])
a1=[["a1","a2","a3"],["b1","b2","b3"],["c1",'c2','c3']]
a2={"fontsize":10,"color":"red"}
sns.heatmap(a,cmap="PuOr")
plt.show()
```
