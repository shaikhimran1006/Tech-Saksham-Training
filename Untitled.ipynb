```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot(df['Fuel_Type'],df['Year'])
```

```python
fuel=df['Fuel_Type'].value_counts()
colors=["green","red","blue"]
fuel.plot(kind='bar',color=colors)
plt.title("fuel consumption")
plt.xlabel("fuel")
plt.ylabel("Cars")
plt.grid()
plt.show()
```

```python
fuel.plot(kind="pie")
plt.legend()
```
